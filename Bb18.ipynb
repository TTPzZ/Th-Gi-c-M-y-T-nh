```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
from scipy import ndimage

image_path='C:/Users/Phuc/TGMTLop6/imgae/dv.jpg'
image = Image.open(image_path)
plt.imshow(image)
```

```python
gray_image = image.convert("L")

image_array =np.array(gray_image)

binary_image = 1*(image_array<200)

labeled_image, num_features = ndimage.label(binary_image)

fig, axs = plt.subplots(3, 1, figsize=(50, 50))

plt.subplot(3,1,1)
plt.imshow(gray_image, cmap='gray')
plt.title("nh xm")

plt.subplot(3,1,2)
plt.imshow(binary_image, cmap='gray')
plt.title("nh nh phn")

plt.subplot(3,1,3)
plt.imshow(labeled_image, cmap='gray')
plt.title(f" nh nhn ({num_features} i tng)")

plt.show();
```

```python
gray_image = image.convert("L")

image_array =np.array(gray_image)

binary_image = 1*(image_array<200)

opened_image = ndimage.binary_closing(binary_image, structure=np.ones((15,15)))

labeled_image, num_features = ndimage.label(opened_image)

fig, axs = plt.subplots(4, 1, figsize=(50, 50))

plt.subplot(4,1,1)
plt.imshow(gray_image, cmap='gray')
plt.title("nh xm")

plt.subplot(4,1,2)
plt.imshow(binary_image, cmap='gray')
plt.title("nh nh phn")

plt.subplot(4,1,3)
plt.imshow(opened_image, cmap='gray')
plt.title("nh sau opening")

plt.subplot(4,1,4)
plt.imshow(labeled_image, cmap='gray')
plt.title(f" nh nhn ({num_features} i tng)")

plt.show();
```
