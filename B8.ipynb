```python
import matplotlib.pyplot as plt
```

```python
x = [1, 2, 3, 4, 5]
y = [10, 11, 12, 13, 15]

x1 = [1, 2, 3, 4, 5]
y1 = [20, 15, 8, 9, 20]

plt.plot(x, y, 'go-', label='Dữ liệu 1')
plt.plot(x1, y1, marker='o', color='y', linestyle='-', label='Dữ liệu 2')
plt.legend()
plt.xlabel('Trục x')
plt.ylabel('Trục y')
plt.show()
```

```python
categories = ['A', 'B', 'C', 'D', 'E']
values = [15, 20, 25, 15, 20]

plt.bar(categories, values, color='g', alpha=0.55)
plt.xlabel('Danh mục')
plt.ylabel('Giá trị')
plt.show()
```

```python
my_labels = ['A', 'B', 'C', 'D']
sizes = [10, 25, 45, 20]
my_colors = ['red', 'green', 'blue', 'yellow']

plt.pie(sizes, labels=my_labels, colors = my_colors, autopct='%1.1f%%') 
plt.title('Biểu đồ hình tròn')
plt.show()
```

```python
x = [1, 2, 3, 4, 5]
y = [10, 11, 12, 13, 15]

plt.scatter(x, y, c='r', marker='o', label='Dữ liệu')
plt.title('Biểu đồ phân tán ví dụ')
plt.xlabel('Trục x') 
plt.ylabel('Trục y') 
plt.legend() 
plt.show() 
```

```python
data = [10, 20, 30, 35, 40, 45, 50, 55, 65]

plt.boxplot(data)
plt.title('Biểu đồ hộp ví dụ') 
plt.ylabel('Giá trị') 
plt.show()
```

```python
data = [10, 20, 30, 35, 40, 45, 50, 55, 65]

plt.violinplot(data)
plt.title('Biểu đồ violin ví dụ') 
plt.ylabel('Giá trị') 
plt.show()
```
